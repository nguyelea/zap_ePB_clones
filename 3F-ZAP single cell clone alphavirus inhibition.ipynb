```python
# Import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Set default figure font to Arial
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'Arial'
rcParams['font.family'] = 'sans-serif'
```

```python
# Import alphavirus fold inhibition values as dataframe
fold_inhibition = pd.read_csv('3fzap_clones_alphavirus_inhibition.csv')
fold_inhibition
```

```python
# Plot inhibition values as catplot
sns.set_style("ticks")
sns.despine()

virus_palette = {'SINV':'#5db100', 'RRV':'#ff786a', 'ONNV':'#00c0c6'}
g = sns.catplot(data=fold_inhibition, 
                kind='bar', 
                x='virus', 
                y='inhibition', 
                col='isoform', 
                alpha=0.6,
                ec='black',
                linewidth=1.5,
                palette=virus_palette)
plt.yscale('log')

# Map individual values to stripplot
g.map(sns.stripplot, 'virus', 'inhibition', order=['SINV', 'RRV', 'ONNV'], palette=virus_palette, alpha=0.8, size=15)

# Change labels
g.set_titles('')
g.set_axis_labels(x_var='', y_var='Fold inhibition\n(relative to -dox)', fontsize=15, weight='bold')
g.set_xticklabels(labels=['SINV','RRV','ONNV'], rotation=45, fontsize=15)
g.tick_params(axis='y', labelsize=12)

# Export plot as PDF
plt.savefig("3F-ZAP clones alphavirus inhibition.pdf", bbox_inches = "tight")
```
